```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

%%matplotlib inline
```

```python
df=pd.read_csv('datasets/data.csv')
```

```python
df
```

```python
data=np.log(df['e_mu_current'])
data_vis=data[9000:10000]

data_std=np.std(data)
data_mean=np.mean(data)

plt.figure(figsize=(15,5))
plt.plot(data_vis)

plt.axhline(data_mean, c='black')
plt.text(data_vis.index[0]-40, data_mean, 'mean', fontsize=20)

plt.axhline(data_std+data_mean, c='yellow')
plt.text(data_vis.index[0]-40, data_std+data_mean, 'std', fontsize=20)

plt.axhline(data_std*2+data_mean, c='red')
plt.text(data_vis.index[0]-40, data_std*2+data_mean, '2std', fontsize=20)

plt.axhline(data_std*3+data_mean, c='green')
plt.text(data_vis.index[0]-40, data_std*3+data_mean, '3std', fontsize=20)

# plt.xlim(0,len(data_vis))
# plt.ylim(-6,-1)
# plt.text
plt.show()
```
